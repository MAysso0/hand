```python
import cv2 as cv                    #importing opencv 
from cv2 import COLOR_RGB2BGR
import mediapipe as mp              #importing mediapipe to detect the hand
import numpy as np
import serial                       #importing serial to communicate with arduino
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
```

```python
def cal_angle_frnt(a,b,c):          #this function
    a = np.array(a) #first joint    #finds the angle between joints
    b = np.array(b) #mid joint
    c = np.array(c) #last joint
    radians = np.arctan2(c[0]-b[0], c[2]-b[2]) - np.arctan2(a[0]-b[0], a[2]-b[2])
    angle = np.abs(radians*180.0/np.pi)
    return angle
```

```python
capture = cv.VideoCapture(0)        #capturing the live footage

# Configure the serial port
ser = serial.Serial('COM3', 9600)   #this line here varies depending on the com number which you can
                                    #find in arduino and the frequency of the serial
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic: #creating the model for the detector

    while True:

        isTrue , frame = capture.read()

        frame = cv.flip(frame,1)    #this line was used to inverse the frame it can be removed depending on the footage provided

        #recolor image to rgb
        frame = cv.cvtColor(frame,cv.COLOR_BGR2RGB)
        frame.flags.writeable = False  #this line saves memory

        #make detection
        results = holistic.process(frame)

        #recoloring back to bgr
        frame.flags.writeable = True
        frame = cv.cvtColor(frame, COLOR_RGB2BGR)

        #time.sleep(0.05)

        try:
            hlandmarks=results.right_hand_landmarks.landmark
            #finger=[hlandmarks[mp_holistic.HandLandmark.INDEX_FINGER_TIP.value].x, hlandmarks[mp_holistic.HandLandmark.INDEX_FINGER_TIP.value].y, hlandmarks[mp_holistic.HandLandmark.INDEX_FINGER_TIP.value].z, hlandmarks[mp_holistic.HandLandmark.INDEX_FINGER_TIP.value].visibility]    #these lines locate some landmarks
            #inner_finger=[hlandmarks[mp_holistic.HandLandmark.WRIST.value].x, hlandmarks[mp_holistic.HandLandmark.WRIST.value].y, hlandmarks[mp_holistic.HandLandmark.WRIST.value].z, hlandmarks[mp_holistic.HandLandmark.WRIST.value].visibility]                                          #they as well can be changed depending on the usage
            #thumb=[hlandmarks[mp_holistic.HandLandmark.THUMB_TIP.value].x, hlandmarks[mp_holistic.HandLandmark.THUMB_TIP.value].y, hlandmarks[mp_holistic.HandLandmark.THUMB_TIP.value].z, hlandmarks[mp_holistic.HandLandmark.THUMB_TIP.value].visibility]
            wrist=[hlandmarks[0].x,hlandmarks[0].y,hlandmarks[0].z,hlandmarks[0].visibility]
            joint1=[hlandmarks[5].x,hlandmarks[5].y,hlandmarks[5].z,hlandmarks[5].visibility]
            joint2=[hlandmarks[6].x,hlandmarks[6].y,hlandmarks[6].z,hlandmarks[6].visibility]
            angle=int(360-cal_angle_frnt(wrist,joint1,joint2)) #in this code I am interested in the angle of the index
            print((str(int(angle))+"\n").encode())
            ser.write((str(int(angle))+"\n").encode())   #sending the value to the arduino
            cv.putText(frame, str(int(angle)), (50,200), cv.FONT_HERSHEY_SIMPLEX, 1,(0, 0, 255),2,cv.LINE_AA)   #printing the value on the screen

        except:
            pass

        mp_drawing.draw_landmarks(frame, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, mp_drawing.DrawingSpec(color=(255,0,0), thickness=2, circle_radius=4), mp_drawing.DrawingSpec(color=(0,225,0), thickness=2, circle_radius=4)) #drawing the right hand

        
        

        cv.imshow("video",frame)

        if cv.waitKey(1) & 0xFF==ord("q"):
            break
        
    ser.close()
    capture.release()
    cv.destroyAllWindows()
```
